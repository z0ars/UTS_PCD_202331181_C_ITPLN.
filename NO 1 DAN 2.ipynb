```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('ALNIZARBAIHAQI.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
plt.title("Citra Kontras")
plt.axis('off')
plt.show()
```

```python
def threshold_warna(image, lower, upper):
    mask = cv2.inRange(image, lower, upper)
    hasil = cv2.bitwise_and(image, image, mask=mask)
    return hasil
```

```python
# Ambang batas tiap warna berdasarkan analisis histogram tadi
lower_red = np.array([160, 0, 0])
upper_red = np.array([220, 150, 150])

lower_green = np.array([0, 140, 0])
upper_green = np.array([150, 200, 150])

lower_blue = np.array([0, 0, 140])
upper_blue = np.array([150, 150, 210])
```

```python
# Hasil deteksi masing-masing warna
red_detected = threshold_warna(img, lower_red, upper_red)
green_detected = threshold_warna(img, lower_green, upper_green)
blue_detected = threshold_warna(img, lower_blue, upper_blue)
```

```python
# Tampilkan hasilnya
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2,2,1)
plt.title("Citra Asli")
plt.imshow(img)
plt.axis("off")

plt.subplot(2,2,2)
plt.title("Deteksi Merah")
plt.imshow(red_detected)
plt.axis("off")

plt.subplot(2,2,3)
plt.title("Deteksi Hijau")
plt.imshow(green_detected)
plt.axis("off")

plt.subplot(2,2,4)
plt.title("Deteksi Biru")
plt.imshow(blue_detected)
plt.axis("off")
```
