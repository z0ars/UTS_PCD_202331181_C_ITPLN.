```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('BACKLIGHT.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(6,4))
plt.imshow(img_rgb)
plt.title('Gambar Asli')
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6,4))
plt.imshow(gray, cmap='gray')
plt.title('Gambar Grayscale')
plt.axis('off')
plt.show()
```

```python
bright = cv2.convertScaleAbs(gray, alpha=1, beta=50)  # beta = brightness
plt.figure(figsize=(6,4))
plt.imshow(bright, cmap='gray')
plt.title('Gambar Gray yang Dicerahkan')
plt.axis('off')
plt.show()
```

```python
contrast = cv2.convertScaleAbs(gray, alpha=2.0, beta=0)  # alpha = contrast
plt.figure(figsize=(6,4))
plt.imshow(contrast, cmap='gray')
plt.title('Gambar Gray yang Diperkuat Kontras')
plt.axis('off')
plt.show()
```

```python
bright_contrast = cv2.convertScaleAbs(gray, alpha=1.5, beta=30)
plt.figure(figsize=(6,4))
plt.imshow(bright_contrast, cmap='gray')
plt.title('Gambar Gray Dicerah & Diperkuat Kontras')
plt.axis('off')
plt.show()
```
